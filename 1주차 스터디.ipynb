2. Numpy

  **1) numpy의 특징**

* 과학 계산을 위한 라이브러리
* 행렬/배열 처리 및 연산
* 난수생성

   **2) numpy의 차원**
* 1차원 축(행) : axis 0 => Vector(벡터)
* 2차원 축(열) : axis 1 => Matrix(행렬)
* 3차원 축(채널) : axis 2 => Tensor(3차원 이상, 텐서)

3. Matplotlib

 **1) Matplotlib의 특징**
* 각각의 함수를 사용하여 간편하게 그래프를 만들고 변화를 줄 수 있다.
  
  예를 들어, 그래프 영역을 만들고, 몇 개의 선을 표현하고, 레이블을 꾸미는 일이 가능하다.

  **2) 직선 그래프 그리는법**
  

```
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.show()
```
위와 같은 코드를 입력하게되면 1차함수 형태의 직선 그래프가 나타난다.
