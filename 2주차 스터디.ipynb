머신러닝 도입
=============

1. 인공지능의 특징
   
* 기본적으로는 학습하고 적용하는 컴퓨터 프로그래밍이다.



2. NLP의 특징과 실용 예시
* NLP는 자연어 처리라고 하며 컴퓨터가 자체적으로 "말하기"를 생성하는 것을 가능하게 한다.
* 컴퓨터가 텍스트를 읽고, 편집하고, 요약하는 데 도움이 된다. 
* NLP를 이용한 예로는 이메일 필터, 텍스트 자동 완성, 언어 번역, 텍스트 분석등 다양한 예시가 있다.

3. CV의 특징과 실용 예시
* CV는 이미지와 비디오를 처리해 유의미한 정보를 추출하는 기술이다.
* CV를 이용한 예로는 문자 인식, 이미지 인식, 패턴 인식, 객체 감지 및 분류 등이 있다.
* CV는 자동화, 프로세스 개선 및 규제 준수와 같은 분야에 많이 활용되고 있다.
